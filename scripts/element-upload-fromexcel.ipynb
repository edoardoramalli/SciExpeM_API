```python
for index, row in df.iterrows():
    spec = my_sciexpem.filterDatabase(model_name='Specie', InChI = row['InChI'])
    if spec == []:
        if not type(row['CAS']) == float:
            cas = row['CAS']
        else:
            cas = None
            
        if not type(row['SMILES']) == float:
            smiles = row['SMILES']
        else:
            smiles = None
        
        if not type(row['Chemical Name']) == float:
            chemname = row['Chemical Name']
        else:
            chemname = None
        alname = []
        if not type(row['Alternative Names']) == float:
            alname.append(row['Alternative Names'])
            alname.append(row['Preferred Name'])
        else:
            alname = None
            
        s1 = Specie(InChI = row['InChI'], 
                    preferredKey = row['Preferred Name'], 
                    formula = row['Formula [CxHyOz]'],
                    CAS=cas,
                    SMILES=smiles,
                    chemName=chemname,
                    names = alname)
        my_sciexpem.insertElement(s1, verbose=True)
```
