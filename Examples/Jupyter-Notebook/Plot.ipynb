## Query the database

```python
experiment_list = connection.filterDatabase(model_name='Experiment', status='verified')

result = {}

for experiment in experiment_list:
    diz = connection.getCurveMatching(experiment=experiment)
    result[experiment.fileDOI] = {key: diz[key]['score'] for key in diz}
```

```python
# This is a support function to plot the result
def df_to_plotly(df):
    return {'z': df.values.tolist(),
            'x': df.columns.tolist(),
            'y': df.index.tolist()}
```

```python
# Plotting
import pandas as pd
import plotly.graph_objects as go

dataframe = pd.DataFrame(result).dropna(axis=1, how='all') # remove void columns

fig = go.Figure(data=go.Heatmap(df_to_plotly(dataframe), colorscale='rdylgn', xgap=2, ygap=2,))
fig['layout'].update(plot_bgcolor='white',
                    title_text='Curve Matching Score. Experiment Vs. Model ', title_x=0.5)
fig.show()
```
